```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
sizes = np.array([600, 700, 800, 900, 1000]).reshape(-1, 1)
prices = np.array([150000, 200000, 250000, 300000, 350000])
```

```python
model = LinearRegression()
model.fit(sizes, prices)
```

```python
predicted_prices = model.predict(sizes)
```

```python
plt.scatter(sizes, prices, color='blue')
plt.plot(sizes, predicted_prices, color='red')
plt.xlabel('Size (sq ft)')
plt.ylabel('Price ($)')
plt.title('House Price Prediction')
plt.show()
```
